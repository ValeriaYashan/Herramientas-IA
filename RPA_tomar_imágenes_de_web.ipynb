```python
#TOMA IMÁGENES DE WEB, LAS MUESTRA, DESCARGA EN CARPETA ZIP
from google.colab import drive
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import os #es una biblioteca estándar que proporciona una interfaz para interactuar con el sistema operativo en el que se ejecuta el programa
from IPython.display import Image, display
#import shutil

# Montar Google Drive
drive.mount('/content/drive')

def get_domain_name(url):
    # Obtener el nombre de dominio desde la URL
    parsed_url = urlparse(url)
    domain_name = parsed_url.netloc.replace(".", "_")
    return domain_name

# URL de la página web con imágenes de remeras
url = 'https://www.clarin.com/'

# Obtener el contenido HTML de la página
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar todas las etiquetas de imagen en la página
img_elements = soup.find_all('img')

# Ruta de la carpeta donde se descargarán las imágenes en Google Drive
download_folder_path = f'/content/drive/MyDrive/downloaded_images_{get_domain_name(url)}'

# Crear la carpeta si no existe
if not os.path.exists(download_folder_path):
    os.makedirs(download_folder_path)

# Descargar cada imagen y guardar la URL en el archivo
for idx, img_element in enumerate(img_elements):
    # Obtener la URL de la imagen, si está disponible
    img_url = img_element.get('src') or img_element.get('data-src')

    if img_url:
        # Concatenar el esquema a las URL relativas
        img_url = urljoin(url, img_url)

        try:
            # Descargar la imagen y manejar redirecciones
            response_img = requests.get(img_url, allow_redirects=True)

            if response_img.status_code == 200 and response_img.content:
                # Guardar la imagen en la carpeta en Google Drive
                img_name = f'image_{idx + 1}.jpg'  # Nombre de la imagen
                img_path = os.path.join(download_folder_path, img_name)
                with open(img_path, 'wb') as img_file:
                    img_file.write(response_img.content)
                print(f"Imagen descargada: {img_path}")

            else:
                print(f"Error al descargar la imagen ({response_img.status_code}): {img_url}")

        except requests.exceptions.MissingSchema:
            print(f"Error: URL de imagen no válida: {img_url}")

print(f"Imágenes descargadas y guardadas en la carpeta '{download_folder_path}'.")

# Comprimir la carpeta en un archivo ZIP en Google Drive
zip_file_path = f"/content/drive/MyDrive/{get_domain_name(url)}.zip"
shutil.make_archive(zip_file_path, 'zip', download_folder_path)

print(f"Carpeta comprimida y archivo ZIP guardado en Google Drive: '{zip_file_path}'.")
```
