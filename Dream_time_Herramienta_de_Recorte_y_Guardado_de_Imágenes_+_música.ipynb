```python
#@title Upload Photo

import os
from google.colab import files
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import interact

# Crear directorio de fotos si no existe
photos_dir = '/content/dreamtime/src/cli/photos/'
if not os.path.exists(photos_dir):
    os.makedirs(photos_dir)

# Solicitar al usuario que cargue una imagen
print('Choose 1 image file (jpg or png)')
uploaded = files.upload()

# Verificar si se carg un archivo de imagen y almacenar su nombre
target = None
for name, bytes_ in uploaded.items():
    if name.endswith('.jpg') or name.endswith('.jpeg'):
        target = name
        break
    elif name.endswith('.png'):
        target = name
        break
    else:
        print('Only image (jpg or png)')

# Procesar la imagen cargada
if target:
    # Abrir la imagen y obtener sus dimensiones
    image = Image.open(target)
    im_w, im_h = image.size

    # Ajustar tamao de la imagen si es necesario y guardarla en el directorio de fotos
    if im_w > 512 or im_h > 512:
        print('Resizing image...')
        image.thumbnail((512, 512))

    # Funcin para dibujar un cuadro en la imagen y guardar la imagen recortada
    def f(BoxSize, x, y, save=False):
        ed_image = image.copy()
        draw = ImageDraw.Draw(ed_image)
        draw.rectangle([(x, y), (x + BoxSize, y + BoxSize)], outline='red', width=10)
        if save:
            print('SAVED!')
            crop_im = image.copy()
            crop_im = crop_im.crop((x, y, x + BoxSize, y + BoxSize))
            crop_im = crop_im.resize((512, 512))
            crop_im.save(photos_dir + target[:-3] + 'png')

        # Mostrar la imagen con el cuadro dibujado
        plt.imshow(np.asarray(ed_image))
        plt.axis('off')
        plt.show()

    # Definir controles deslizantes y botn para guardar la imagen recortada
    def on_butt_clicked(event):
        f(w_box.value, w_x.value, w_y.value, save=True)

    w_box = widgets.IntSlider(min=100, max=min(im_w, im_h), value=512)
    w_x = widgets.IntSlider(min=0, max=im_w - 512, value=0)
    w_y = widgets.IntSlider(min=0, max=im_h - 512, value=0)
    butt = widgets.Button(description='Save')
    butt.on_click(on_butt_clicked)

    # Mostrar botn y controles deslizantes interactivos
    display(butt)
    interact(f, BoxSize=w_box, x=w_x, y=w_y)
```

```python
import music21

# Crear una lista de las notas musicales
notas = ['C', 'D', 'E', 'F', 'G', 'A', 'B']

# Crear una lista para almacenar las notas musicales creadas
notas_creadas = []

# Crear una nota musical para cada nota en la lista
for nota in notas:
    # Crear una nota con octava 4
    nueva_nota = music21.note.Note(nota + '4')
    # Agregar la nueva nota a la lista
    notas_creadas.append(nueva_nota)

# Crear una partitura y agregar las notas
partitura = music21.stream.Stream()
partitura.append(notas_creadas)

# Reproducir la partitura en formato MIDI
partitura.show('midi')
```
